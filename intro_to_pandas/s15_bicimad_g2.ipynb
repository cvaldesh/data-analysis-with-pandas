### Creating time-related variables
After converting "date" column from string to datetime

```python
bicimad["year"] = bicimad["date"].map(lambda x: x.year)
bicimad["month"] = bicimad["date"].dt.month
bicimad["weekday"] = bicimad["date"].dt.weekday
bicimad["weekend"] = np.where(
   bicimad["weekday"]<5,
    0,
    1
)

bicimad
```

```python
def get_season(date_to_convert: datetime.date):
    date_year = date_to_convert.year
    seasons = [
        ("winter", date(date_year, 12, 21), date(date_year, 12, 31)),
        ("spring", date(date_year, 3, 20), date(date_year, 6, 21)),
        ("summer", date(date_year, 6, 22), date(date_year, 9, 21)),
        ("fall", date(date_year, 9, 22), date(date_year, 12, 20)),
        ("winter", date(date_year, 1, 1), date(date_year, 3, 19))
    ]
    
    for elem in seasons:  # season = tuple(name, start, end)
        if date_to_convert>=elem[1] and date_to_convert<=elem[2]:
            return elem[0]
        
```
